```python
import numpy as np
import pandas as pd 
import random
import math
random.seed(111)
```

```python
dates = ["2020/04/07", "2020/04/21", "2020/05/05", "2020/05/19", "2020/06/02"]
dates = ['2020/03/24']
```

```python
data = pd.read_csv('../../init_data.csv', header=0, index_col=0)
data
```

```python
S = 51
T = 14
def sto_file_generator(date):
    #open the cor file
    date = ('_').join(date.split('/'))
    prob_name = 'SR_{}'.format(date)
    cor_file = open(prob_name + '.sto', 'w')
    
    cor_file.write('STOCH ' + prob_name + ' \n')
    cor_file.write('BLOCKS DISCRETE REPLACE\n')
    probability = 1 / 3
    
    
    for j in range(S):
        demand_state = pd.read_csv('../../IHME_data/' + date + '/' + data.loc[j, 'State'] + '.csv', header=0, index_col=0)
        paths = [demand_state['InvVen_mean'], demand_state['InvVen_mean'] + (demand_state['InvVen_upper'] - demand_state['InvVen_mean']) / 2, demand_state['InvVen_mean'] + (demand_state['InvVen_lower'] - demand_state['InvVen_mean']) / 2] 
        for k in range(3):
            cor_file.write(' BL BLOCK{} PERIOD2 {}\n'.format(j, probability))

            path = {}
            for time in range(1, T + 1):
                n = 'c_e_dummy_d({}_{})_'.format(j, time)
                path[n] =  math.ceil(paths[k][time - 1])
            path_sorted = sorted(path.items())
            for i in path_sorted:
                cor_file.write('\t RHS {} {}\n'.format(i[0], i[1]))

       
    cor_file.write('ENDATA \n')
    cor_file.close()
    
```

```python
for i in dates:
    sto_file_generator(i)
```

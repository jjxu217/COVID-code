```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
import math
```

```python
ini = pd.read_csv('../init_data.csv', header=0, index_col=0)
ini.loc[51, 'State'] = 'SNS'
#general param
states_name = ini['State']
first_date = '2020/03/25'
P = 10

new_first_date = '2020/11/19'
new_P = 4

t = 7
T = 14
P = 10
S = 51

init_ratio = 0.6
```

# No allocation

```python
no_action = pd.read_csv('../IHME/0602_eval/Alaska.csv', header=0, index_col=0)['InvVen_mean']
no_action.index = pd.to_datetime(no_action.index)
no_action = no_action[no_action.index <= pd.to_datetime("2020/06/02")]
no_action[no_action>0] = 0

res = pd.DataFrame(index=no_action.index)

for i, name in enumerate(ini['State'][:-1]):
    state_data = pd.read_csv('../IHME/0602_eval/{}.csv'.format(name), header=0, index_col=0)['InvVen_mean']
    state_data.index = pd.to_datetime(state_data.index)
    state_data = state_data[state_data.index <= pd.to_datetime("2020/06/02")]

    unmet = state_data - init_ratio * ini.loc[i, 'Available capacity']
    unmet[unmet < 0] = 0
    no_action = unmet + no_action
    if sum(unmet) > 0:
        res[name] = unmet

res.plot.line()
```

```python
def performance(param):
    models = ['det_no_SNS', 'det', 'SR', 'GR', 'GRA']
    temp = pd.read_csv('../IHME/0602_eval/Alaska.csv', header=0, index_col=0)['InvVen_mean']
    res_all = pd.DataFrame(index=temp.index)
    res_all['no allocation'] = no_action
    for model in models:
        sol = [{} for _ in range(P + new_P)]
        for i in range(P):
            date = pd.to_datetime(first_date) + timedelta(i * t)
            sol_dt = pd.read_csv('res_all/{}/{}/{}.csv'.format(param, model, date.strftime('%m%d')), header=0)
            for r in range(S):  
                sol[i][sol_dt.loc[r, 'State']] = sol_dt.loc[r, 'stock_self'] + sum(eval(sol_dt.loc[r, 'stock_nb']))
            sol[i]['SNS'] = sol_dt.loc[51, 'stock_self']
        for i in range(new_P):
            date = pd.to_datetime(new_first_date) + timedelta(i * t)
            sol_dt = pd.read_csv('res_all/{}/{}/{}.csv'.format(param, model, date.strftime('%m%d')), header=0)
            for r in range(S):  
                sol[P + i][sol_dt.loc[r, 'State']] = sol_dt.loc[r, 'stock_self'] + sum(eval(sol_dt.loc[r, 'stock_nb']))
            sol[P + i]['SNS'] = sol_dt.loc[51, 'stock_self']
        stock = pd.DataFrame(index=list(ini['State']))

        stock['init'] = [round(init_ratio*i) for i in ini['Available capacity']]
        for i in range(P):
            date = pd.to_datetime(first_date) + timedelta(i * t)
            stock[date.strftime('%m%d')] = [sol[i][k] for k in list(ini['State'])]
        
        for i in range(new_P):
            date = pd.to_datetime(new_first_date) + timedelta(i * t)
            stock[date.strftime('%m%d')] = [sol[i + P][k] for k in list(ini['State'])]
        stock.to_csv('res_all/{}/{}_allstock.csv'.format(param, model))
    
        #before 06/02
        model_eval = pd.read_csv('../IHME/0602_eval/Alaska.csv', header=0, index_col=0)['InvVen_mean']
        model_eval.index = pd.to_datetime(model_eval.index)
        model_eval = model_eval[model_eval.index <= pd.to_datetime("2020/06/02")]
        model_eval[model_eval>0] = 0
        res = pd.DataFrame(index=model_eval.index)

        for i, name in enumerate(ini['State'][:S]):
            state_data = pd.read_csv('../IHME/0602_eval/' + name + '.csv', header=0, index_col=0)['InvVen_mean']
            state_data.index = pd.to_datetime(state_data.index)
            state_data = state_data[state_data.index <= pd.to_datetime("2020/06/02")]

            for time_slot in range(P):
                state_data.iloc[t * time_slot : t * (time_slot + 1)] = state_data.iloc[t * time_slot : t * (time_slot + 1)] - sol[time_slot][name]
            state_data[state_data < 0] = 0
            model_eval = state_data + model_eval

            if sum(state_data) > 0:
                res[name] = state_data
        if not res.empty:
            res.plot.line()
        res_all[model] = model_eval
        
    res_all.plot.line()
    for model in models:
        print('Total unmet for {} is {}'.format(model, sum(res_all[model])))
    
    
```

```python
def performance(param):
    no_action = pd.read_csv('../IHME/0602_eval/Alaska.csv', header=0, index_col=0)['InvVen_mean']
    no_action1 = pd.read_csv('../IHME/1217_eval/Alaska.csv', header=0, index_col=0)['InvVen_mean']
    no_action = pd.concat([no_action, no_action1])
    no_action.index = pd.to_datetime(no_action.index)
    no_action[no_action>0] = 0

    res = pd.DataFrame(index=no_action.index)

    for i, name in enumerate(ini['State'][:-1]):
        state_data = pd.read_csv('../IHME/0602_eval/{}.csv'.format(name), header=0, index_col=0)['InvVen_mean']
        state_data1 = pd.read_csv('../IHME/1217_eval/{}.csv'.format(name), header=0, index_col=0)['InvVen_mean']
        state_data = pd.concat([state_data, state_data1])
        state_data.index = pd.to_datetime(state_data.index)

        unmet = state_data - init_ratio * ini.loc[i, 'Available capacity']
        unmet[unmet < 0] = 0
        no_action = unmet + no_action
        if sum(unmet) > 0:
            res[name] = unmet


    models = ['det', 'SR', 'GR', 'GRA']
#     temp = pd.read_csv('../IHME/0602_eval/Alaska.csv', header=0, index_col=0)['InvVen_mean']
#     temp1 = pd.read_csv('../IHME/1217_eval/Alaska.csv', header=0, index_col=0)['InvVen_mean']
#     temp = pd.concat([temp, temp1])
    res_all = pd.DataFrame(index=no_action.index)
    res_all['no allocation'] = no_action
    for model in models:
        sol = [{} for _ in range(P + new_P)]
        for i in range(P):
            date = pd.to_datetime(first_date) + timedelta(i * t)
            sol_dt = pd.read_csv('res_all/{}/{}/{}.csv'.format(param, model, date.strftime('%m%d')), header=0)
            for r in range(S):  
                sol[i][sol_dt.loc[r, 'State']] = sol_dt.loc[r, 'stock_self'] + sum(eval(sol_dt.loc[r, 'stock_nb']))
            sol[i]['SNS'] = sol_dt.loc[51, 'stock_self']
        for i in range(new_P):
            date = pd.to_datetime(new_first_date) + timedelta(i * t)
            sol_dt = pd.read_csv('res_all/{}/{}/{}.csv'.format(param, model, date.strftime('%m%d')), header=0)
            for r in range(S):  
                sol[P + i][sol_dt.loc[r, 'State']] = sol_dt.loc[r, 'stock_self'] + sum(eval(sol_dt.loc[r, 'stock_nb']))
            sol[P + i]['SNS'] = sol_dt.loc[51, 'stock_self']
        stock = pd.DataFrame(index=list(ini['State']))

        stock['init'] = [round(init_ratio*i) for i in ini['Available capacity']]
        for i in range(P):
            date = pd.to_datetime(first_date) + timedelta(i * t)
            stock[date.strftime('%m%d')] = [sol[i][k] for k in list(ini['State'])]
        
        for i in range(new_P):
            date = pd.to_datetime(new_first_date) + timedelta(i * t)
            stock[date.strftime('%m%d')] = [sol[i + P][k] for k in list(ini['State'])]
        stock.to_csv('res_all/{}/{}_allstock.csv'.format(param, model))
    
        #before 06/02
        model_eval = no_action.copy(deep=True)
        model_eval[model_eval>0] = 0
        res = pd.DataFrame(index=model_eval.index)

        for i, name in enumerate(ini['State'][:S]):
            state_data = pd.read_csv('../IHME/0602_eval/' + name + '.csv', header=0, index_col=0)['InvVen_mean']
            state_data1 = pd.read_csv('../IHME/1217_eval/' + name + '.csv', header=0, index_col=0)['InvVen_mean']
            state_data = pd.concat([state_data, state_data1])
            state_data.index = pd.to_datetime(state_data.index)
          

            for i in range(P):
                state_data.iloc[t * i : t * (i + 1)] = state_data.iloc[t * i: t * (i + 1)] - sol[i][name]
            for i in range(P, P + new_P):
                state_data.iloc[t * i : t * (i + 1)] = state_data.iloc[t * i: t * (i + 1)] - sol[i][name]
            state_data[state_data < 0] = 0
            model_eval = state_data + model_eval

            if sum(state_data) > 0:
                res[name] = state_data
        if not res.empty:
            res.plot.line()
        res_all[model] = model_eval
        
    res_all.plot.line()
    return res_all.astype(int)
```
